# Context
Bob has started his own mobile company. 
He wants to give tough fight to big companies like Apple,Samsung etc.
He does not know how to estimate price of mobiles his company creates. 
In this competitive mobile phone market you cannot simply assume things. 
To solve this problem he collects sales data of mobile phones of various companies.

1. Bob wants to find out some relation between features of a mobile phone(eg:- RAM,Internal Memory etc) and its selling price. But he is not so good at Machine Learning. So he needs your help to solve this problem.

2. In this problem you do not have to predict actual price but a price range indicating how high the price is

## Feature Engineering and Extraction.

```python
df.columns
```

1. Separating Independent <b>(x)</b> and Dependent <b>(y)</b> variables.
2. Taking only necessary Independent variables for model.

```python
x=df_train[['ram','battery_power','px_width']]
x
```

```python
y=df_train[['price_range']]
y
```

Using Random Forest Classifier for model creation.

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier()
model.fit(x,y)
```

Crossvalidation

```python
from sklearn.model_selection import cross_validate
```

```python
CrossVal=cross_validate(model,x,y,cv=5,return_train_score=True)
print('Train Score Value : ', CrossVal['train_score'])
print('Test Score Value : ', CrossVal['test_score'])
print('Fit Time : ', CrossVal['fit_time'])
print('Score Time : ', CrossVal['score_time'])
```

<b>Cross Validation Of Model</b>
1. Train Score Value :  [1. 1. 1. 1. 1.]
2. Test Score Value :  [0.8675 0.885  0.8975 0.85   0.865 ]
3. Fit Time :  [0.17755127 0.16755223 0.16655684 0.1835351  0.16755223]
4. Score Time :  [0.01296282 0.0129652  0.0179255  0.01396775 0.01296544]

```python
Test_Score_Value =[0.8675, 0.885, 0.8975, 0.85, 0.865 ]
np.mean(Test_Score_Value)
```

### Model Cross Validation Test Score= 0.873

```python
x_test=df_test[['ram','battery_power','px_width']]
```

```python
pred=model.predict(x_test)
```

```python
pred.shape
```

```python
pred=pd.DataFrame(pred)
```

```python
df_test['Prediction']=pred[0]
```
